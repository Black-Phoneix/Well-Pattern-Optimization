```python
import json

# 定义要修复的文件名
input_file = 'main.ipynb'
output_file = 'main_fixed.ipynb'

try:
    # 1. 以文本方式读取文件
    with open(input_file, 'r', encoding='utf-8') as f:
        content = f.read()
    
    # 2. 替换常见的错误引号
    # 替换中文/智能双引号
    content = content.replace('“', '"').replace('”', '"')
    # 替换中文/智能单引号
    content = content.replace('‘', "'").replace('’', "'")
    
    # 3. 确保文件以 { 开头（去除开头的非法字符）
    start_index = content.find('{')
    if start_index != -1:
        content = content[start_index:]
    
    # 4. 尝试解析一下看是否修好了
    try:
        json_obj = json.loads(content)
        print("格式验证成功！")
        
        # 5. 保存修复后的文件
        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(json_obj, f, indent=1, ensure_ascii=False)
        print(f"修复完成！请在 Jupyter Notebook 中打开: {output_file}")
        
    except json.JSONDecodeError as e:
        print(f"自动修复失败，可能还有其他格式问题: {e}")
        # 仍然写入文件以便手动检查
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write(content)

except FileNotFoundError:
    print(f"找不到文件: {input_file}，请确保脚本和笔记本在同一目录下。")
```
